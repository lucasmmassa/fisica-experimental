# Plot dos gráficos dos resistores

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(100, 100))
```

### Resistor de 47ohm

```python
data_47 = pd.read_excel('47ohm.xlsx')
data_100 = pd.read_excel('100ohm.xlsx')
data_200 = pd.read_excel('200ohm.xlsx')
data_2k = pd.read_excel('2Kohm.xlsx')
data_560k = pd.read_excel('560Kohm.xlsx')
```

```python
x1 = data_47['Corrente']
y1 = data_47['Tensão']

plt.subplot(5,1,1)
plt.plot(x1,y1)
plt.xlabel('Corrente (mA)')
plt.ylabel('Tensão (V)')
plt.title('Resistor de 47 Ohm - Fundo de escala 200mA')

x2 = data_100['Corrente']
y2 = data_100['Tensão']

plt.subplot(5,1,2)
plt.plot(x2,y2)
plt.xlabel('Corrente (mA)')
plt.ylabel('Tensão (V)')
plt.title('Resistor de 100 Ohm - Fundo de escala 200mA')

x3 = data_200['Corrente']
y3 = data_200['Tensão']

plt.subplot(5,1,3)
plt.plot(x3,y3)
plt.xlabel('Corrente (mA)')
plt.ylabel('Tensão (V)')
plt.title('Resistor de 200 Ohm - Fundo de escala 200mA')

x4 = data_2k['Corrente']
y4 = data_2k['Tensão']

plt.subplot(5,1,4)
plt.plot(x4,y4)
plt.xlabel('Corrente (mA)')
plt.ylabel('Tensão (V)')
plt.title('Resistor de 2 KOhm - Fundo de escala 20mA')

x5 = data_560k['Corrente']
y5 = data_560k['Tensão']

plt.subplot(5,1,5)
plt.plot(x5,y5)
plt.xlabel('Corrente (mA)')
plt.ylabel('Tensão (V)')
plt.title('Resistor de 560 KOhm - Fundo de escala 200uA')
```
