```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.polynomial.polynomial import polyfit
```

## Lendo dados das planilhas

```python
df1 = pd.read_csv('dados1.csv')
```

```python
df2 = pd.read_csv('dados2.csv')
```

```python
df3 = pd.read_csv('dados3.csv')
```

### Calculando a força magnética em cada caso

```python
def calular_fm(df,m):
    fm = []
    pi = (m/1000)*9.8
    for index, row in df.iterrows():
        pf = (row['massa']/1000)*9.8
        delta = pf-pi
        fm.append(delta)
    
    return fm
```

#### Caso 1

```python
m1_inicial = df1['massa'].iloc[0]
m1_inicial
```

```python
fm1 = calular_fm(df1,m1_inicial)
```

```python
df1['força'] = fm1
```

```python
df1
```

```python
plt.figure(figsize=(10,10))
plt.title('Força Magnética - L = 12,5mm e n = 1')
plt.scatter(df1['corrente'],df1['força'],color='red',lw=5, label='Medições')
plt.ylim(top=0.006, bottom=-0.0001)
b, a = polyfit(df1['corrente'],df1['força'],1)
plt.plot(df1['corrente'],a*df1['corrente']+b,color='black',lw=2,label='Regressão linear - coeficiente angular: '+str(a))
plt.legend(loc='lower right')
B1 = a/0.0125
plt.savefig('fm1.png')
```
