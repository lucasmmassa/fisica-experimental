# Plot dos gráficos dos resistores

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data_47 = pd.read_excel('47ohm.xlsx')
data_100 = pd.read_excel('100ohm.xlsx')
data_200 = pd.read_excel('200ohm.xlsx')
data_2k = pd.read_excel('2Kohm.xlsx')
data_560k = pd.read_excel('560Kohm.xlsx')
```

```python
l1 = LinearRegression()
l2 = LinearRegression()
l3 = LinearRegression()
l4 = LinearRegression()
l5 = LinearRegression()
```

```python
fig = plt.figure(figsize=(10,5))
plt.tight_layout

x1 = np.array(data_47['Corrente']).reshape(1,-1)
y1 = np.array(data_47['Tensão']).reshape(1,-1)
l1.fit(x1,y1)
p1 = l1.predict(x1)

ax1 = fig.add_axes([0.1, 0.1, 0.1, 0.25])
ax2 = fig.add_axes([0.4, 0.1, 0.1, 0.25])
ax3 = fig.add_axes([0.7, 0.1, 0.1, 0.25])
ax4 = fig.add_axes([0.1, 0.6, 0.1, 0.25])
ax5 = fig.add_axes([0.5, 0.6, 0.1, 0.25])

ax1.plot(x1,y1)
ax1.set_xlabel('Corrente (mA)')
ax1.set_ylabel('Tensão (V)')
ax1.set_title('Gráfico 8 - Resistor de 47 Ohm\nFundo de escala 200mA')

x2 = np.array(data_100['Corrente']).reshape(1,-1)
y2 = np.array(data_100['Tensão']).reshape(1,-1)
l2.fit(x2,y2)
p2 = l2.predict(x2)

ax2.plot(x2,y2,color='r')
ax2.set_xlabel('Corrente (mA)')
ax2.set_ylabel('Tensão (V)')
ax2.set_title('Gráfico 9 - Resistor de 100 Ohm\nFundo de escala 200mA')


x3 = np.array(data_200['Corrente']).reshape(1,-1)
y3 = np.array(data_200['Tensão']).reshape(1,-1)
l3.fit(x3,y3)
p3 = l3.predict(x3)

ax3.plot(x3,y3,color='g')
ax3.set_xlabel('Corrente (mA)')
ax3.set_ylabel('Tensão (V)')
ax3.set_title('Gráfico 10 - Resistor de 200 Ohm\nFundo de escala 200mA')

x4 = np.array(data_2k['Corrente']).reshape(1,-1)
y4 = np.array(data_2k['Tensão']).reshape(1,-1)
l4.fit(x4,y4)
p4 = l4.predict(x4)

ax4.plot(x4,y4,color='y')
ax4.set_xlabel('Corrente (mA)')
ax4.set_ylabel('Tensão (V)')
ax4.set_title('Gráfico 6 - Resistor de 2 KOhm\nFundo de escala 20mA')

x5 = np.array(data_560k['Corrente']).reshape(1,-1)
y5 = np.array(data_560k['Tensão']).reshape(1,-1)
l5.fit(x5,y5)
p5 = l5.predict(x5)

ax5.plot(x5,y5,color='b')
ax5.set_xlabel('Corrente (mA)')
ax5.set_ylabel('Tensão (V)')
ax5.set_title('Gráfico 7 - Resistor de 560 KOhm\nFundo de escala 200uA')

plt.savefig('ajustes.png')
```
